### KNIME 기본 사용법

```python
#Data Table 불러오기 및 확인
import pandas as pd
from IPython.display import display,HTML
df=pd.read_csv("boston_missing.csv")
display(HTML(df.to_html()))
```

```python
#데이터 검토
def skimming(df):
    print("<<DESCRIBE>>")
    print(df.describe())
    print("\n<<SKEWNESS>>")
    print(df.skew())
    print("\n<<KURTOSIS>>")
    print(df.kurtosis())
    print("\n<<MISSING>>")
    print(df.isna().sum())
    print("\n<<HISTOGRAM>>")
    df.hist()
skimming(df)
```

```python
#결측값 대체
#CHAS에 6개의 MISSING VALUE발견, CHAS는 명목변수
#최빈값으로 대체


df1 = df.fillna(df.mode()["CHAS"][0])
df1
```

```python
#결측값 삭제
df2 = df.dropna()
df2.isna().sum()
len(df2)
```

# 실습 2-4
### KNIME을 활용한 데이터 전처리  - Feature creation and Data transformation

```python
#데이터 불러오기
import pandas as pd
from IPython.display import display, HTML

df = pd.read_csv('통신.txt',sep=" ", engine='python')
display(HTML(df.to_html()))
```

```python
skimming(df)
```

```python
#가입 통신사별 비율 파이차트로 확인하기
from matplotlib.pyplot import pie, axis, show

sums = df.groupby(df.가입통신사).sum()
axis('equal')
pie(sums, labels=sums.index)
show()
```

```python
#가입통신사에서 공백을 제거한 새로운 변수 "통신사_공백제거" 생성
df["통신사_공백제거"]=df["가입통신사"].str.replace(" ","")
display(HTML(df.to_html()))
```

```python
#가입통신사에서 공백을 제거해 기존 변수 대체

df["가입통신사"]=df["가입통신사"].str.replace(" ","")
df
sums = df.groupby(df.가입통신사).sum()
axis('equal')
pie(sums, labels=sums.index)
show()
```

```python
#주민번호에서 성별/나이 추출해 변수생성
df["성별"]=df["주민번호"].apply(lambda x : "M" if "-1" in x else "F")
df["나이"]=df["주민번호"].apply(lambda x : 2019-(1900+int(x[0:2])))
```

```python
#전체 데이터 필드 순서 변경
cols = ['ID', '주민번호', '성별', '나이', '직종', '가입통신사', '통화료', '통신사_공백제거']
df=df[cols]
df
```

```python
#데이터의 통계적 특성 및 상관관계 파악하기
import seaborn as sns

corr_matrix=df.corr()

sns.heatmap(corr_matrix, xticklabels=corr_matrix.columns, yticklabels=corr_matrix.columns)
print(corr_matrix)

skimming(df)
```
